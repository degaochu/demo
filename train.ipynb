```python
from autogluon.tabular import TabularDataset, TabularPredictor
from sklearn.model_selection import train_test_split
import pandas as pd


def train(parameter="pm25", presets='medium_quality'):
    #train_data = TabularDataset(train_file)
    train_file=f'train_{parameter}.csv'
    df = pd.read_csv(train_file)
    train_data, test_data = train_test_split(df, test_size=0.2, random_state=1)
    label = 'value'
    predictor = TabularPredictor(label=label, path=f'aqi_model_{parameter}_{presets}').fit(train_data, presets=presets) #presets='', 'good_quality', 'high_quality','best_quality'
    
    y_pred = predictor.predict(test_data.drop(columns=[label]))
    print(y_pred.head())
    print(predictor.evaluate(test_data))
    
    print(predictor.leaderboard(test_data))

    print(predictor.feature_importance(test_data))

    print(f"models: {predictor.model_names()}")
    print(f"best model: {predictor.model_best}")

    
```

```python
train(parameter='o3')
```

```python
train(parameter='pm25')
```

```python
train()
```
